```
import matplotlib.pyplot as plt
import numpy as np
import json
```

# Plot correlation matrix of VpA data

```
with open('./alephData/alephVpA.json') as json_file:
    data = json.load(json_file)['data']

print(data['sbin'])
```

```
corMat = np.zeros((80, 80))
for i in range(1, 81):
    corMat[i-1] = data['corerr{:02d}'.format(i)]
```

```
cax = plt.matshow(corMat)
plt.gcf().set_size_inches(12, 7)
plt.gcf().colorbar(cax)
plt.show()
```
