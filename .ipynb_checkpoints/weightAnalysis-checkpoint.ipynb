# Monomial weights

```python
x = np.arange(0, mTau**2, 0.01)
plt.plot(x, list(map(lambda x: (1-x/mTau**2),x)), label=r'$(1-s/m_\tau^2)$')
plt.plot(x, list(map(lambda x: (1-x/mTau**2)**2,x)), label=r'$(1-s/m_\tau^2)^2$')
plt.plot(x, list(map(lambda x: (1-x/mTau**2)**3,x)), label=r'$(1-s/m_\tau^2)^3$')
plt.plot(x, list(map(lambda x: (1-x/mTau**2)**4,x)), label=r'$(1-s/m_\tau^2)^4$')
plt.legend(prop={'size': 20})

plt.xlabel('s in GeV', fontsize=22)
plt.ylabel(r'$(1-s/m_\tau^2)^k$', fontsize=22)

ax = plt.gca()
ax.tick_params(axis='both', labelsize=18)

fig = plt.gcf()
fig.set_size_inches(12, 7)
fig.patch.set_facecolor('white')
fig.tight_layout()
plt.savefig('./../Thesis/chapters/tauDecaysIntoHadrons/images/monomialWeightGraphs.eps',format='eps')
```
