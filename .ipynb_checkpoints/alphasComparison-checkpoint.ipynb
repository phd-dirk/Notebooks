```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = [
    # {
    # 'name': r'2005 - \textsc{aleph}',
    # 'cite': 'Aleph2005',
    # 'aTau': 0.340,
    # 'aTauErr': (0.015, 0.015),
    # # 'aZ': 0.1209,
    # # 'aZErr': (0.0018,0.0018)
    # },
    {
    'name': r'2008 - Beneke, Jamin',
    'cite': 'Beneke2008',
    'aTau': 0.320,
    'aTauErr': (0.012,-0.007),
}, {
    'name': r'2014 - Davier, Hoecker, Malaescu, Yuan, Zhang',
    'cite': 'Davier2014',
    'aTau': 0.332,
    'aTauErr': (0.012,0.012),
}, {
    'name': r'2016 - Pich, Rodríguez-Sánchez',
    'cite': 'Pich2016',
    'aTau': 0.328,
    'aTauErr': (0.012,0.012),
}, {
    'name': r'2016 - Boito, Golterman, Jamin, Maltman, Osborne, Peris',
    'cite': '',
    'aTau': 0.301,
    'aTauErr': (0.01,0.01),
}, {
    'name': r'PDG',
    'cite': 'PDG2018',
    'aTau': 0.325,
    'aTauErr': (0.015,0.015),
}]

x = list(map(lambda alpha: alpha['aTau'], data))
y = list(map(lambda alpha: alpha['name'], data))
# y = range(0, 5)
x_err = list(map(lambda alpha: alpha['aTauErr'][0], data))
x_mean = [np.mean(x)]*len(x)


print(x_mean)
plt.errorbar(x, y, xerr=x_err, fmt='o', label=r'$\alpha_s$')
plt.plot(x_mean, y, linestyle='--', label='mean')
plt.grid()
plt.legend()

fig = plt.gcf()
# fig.set_size_inches(14, 10)
fig.patch.set_facecolor('white')

plt.savefig('../Thesis/chapters/measuringTheStrongCoupling/images/historicAlphasComparison.eps', type='eps')
```
