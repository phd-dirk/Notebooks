In this plot we compare the $c_6$ and $c_8$ OPE coefficients of the best fits by $\chi^2/dof$ value. Even if the coefficients show a huge variation of value we present their mean and variance: $\mu_{c_6}=-0.252$ with $\sigma^2_{c_6}=0.29$ and $\mu_{c_8}=0.077$ with $\sigma^2_{c_8}=2.01$. We see that $c_6$ takes a negative and $c_{8}$ a postive value for most of the fits. Furthermore due to the variance of the OPE coefficients it is dificult to choose the right value for them. As there are more problems with including higher dimension contributions of the OPE we do not include them in our comparison.
