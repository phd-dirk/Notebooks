```python
import numpy as np
import pandas as pd

wCubeD6D8_df = pd.read_csv('../csv/wCubeD6D8.csv', skipfooter=1, engine='python')
wCubeD6D8_df['label']=[
    'wCube_D6D8_smin1975', 'wCube_D6D8_smin2050', 'wCube_D6D8_smin2150',
    'wCube_D6D8_smin2250', 'wCube_D6D8_smin2350', 'wCube_D6D8_smin2500'
]

wCubeD6D8Alpha_df = pd.read_csv('../csv/wCubeD6D8Alpha.csv', skipfooter=1, engine='python')
wCubeD6D8Alpha_df['label']=[
    'wCube_D6D8Alpha_smin1975', 'wCube_D6D8Alpha_smin2050', 'wCube_D6D8Alpha_smin2150',
    'wCube_D6D8Alpha_smin2250'
]

wCubeD6D8D10_df = pd.read_csv('../csv/wCubeD6D8D10.csv', skipfooter=1, engine='python')
wCubeD6D8D10_df['label']=[
    'wCube_D6D8D10_smin1975', 'wCube_D6D8D10_smin2050', 'wCube_D6D8D10_smin2150',
    'wCube_D6D8D10_smin2250', 'wCube_D6D8D10_smin2350'
]

wCubeD6D8D10D12_df = pd.read_csv('../csv/wCubeD6D8D10D12.csv')
wCubeD6D8D10D12_df['label']=[
    'wCube_D6D8D10D12_smin1975', 'wCube_D6D8D10D12_smin2050', 'wCube_D6D8D10D12_smin2150',
    'wCube_D6D8D10D12_smin2250'
]

wKinematicD6D8_df = pd.read_csv('../csv/wKinematicD6D8.csv', skipfooter=1, engine='python')
wKinematicD6D8_df['label']=[
    'wKinematic_D6D8_smin1975', 'wKinematic_D6D8_smin2050', 'wKinematic_D6D8_smin2150',
    'wKinematic_D6D8_smin2250', 'wKinematic_D6D8_smin2350', 'wKinematic_D6D8_smin2500'
]

wKinematicD6D8Alpha_df = pd.read_csv('../csv/wKinematicD6D8Alpha.csv', skipfooter=1, engine='python')
wKinematicD6D8Alpha_df['label']=[
    'wKinematic_D6D8Alpha_smin1975', 'wKinematic_D6D8Alpha_smin2050', 'wKinematic_D6D8Alpha_smin2150',
    'wKinematic_D6D8Alpha_smin2250', 'wKinematic_D6D8Alpha_smin2350'
]

D6_df = pd.concat([wCubeD6D8_df, wCubeD6D8Alpha_df, wCubeD6D8D10_df, wCubeD6D8D10D12_df, wKinematicD6D8_df, wKinematicD6D8Alpha_df], sort=False)
D6_df.set_index('label', inplace=True)
D8_df = pd.concat([wCubeD6D8_df, wCubeD6D8Alpha_df, wCubeD6D8D10_df, wCubeD6D8D10D12_df, wKinematicD6D8_df, wKinematicD6D8Alpha_df], sort=False)
D8_df.set_index('label', inplace=True)
Alpha_df = pd.concat([wCubeD6D8Alpha_df, wKinematicD6D8Alpha_df], sort=False)
Alpha_df.set_index('label', inplace=True)
```
