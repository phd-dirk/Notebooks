```python
# Import and cobine best rows
alD6D8Fits = ['wKinAlD6D8', 'wCubeAlD6D8', 'wQuarticAlD6D8', 'wMonoX3AlD8', 'wMonoX4AlD10', 'wOpt20AlD4D6', 'wOpt30AlD4D6D8']
alphaFits = ['wKinAlD6D8', 'wCubeAlD6D8', 'wCubeAlD6D8D10', 'wQuarticAlD6D8',
             'wQuarticAlD6D8D10', 'wQuarticAlD6D8D10D12', 'wMonoX3AlD8', 'wMonoX4AlD10',
             'wOpt20AlD4D6', 'wOpt30AlD4D6D8']
D6D8Fits = ['wKinAlD6D8', 'wCubeAlD6D8', 'wCubeAlD6D8D10', 'wQuarticAlD6D8',
            'wQuarticAlD6D8D10', 'wQuarticAlD6D8D10D12', 'wKinD6D8', 'wCubeD6D8',
            'wCubeD6D8D10', 'wQuarticD6D8', 'wQuarticD6D8D10', 'wQuarticD6D8D10D12',
            'wOpt30AlD4D6D8']
D10Fits = ['wCubeAlD6D8D10', 'wQuarticAlD6D8D10', 'wQuarticAlD6D8D10D12', 'wMonoX4AlD10']

df_alD6D8 = pd.DataFrame()
for fit in alD6D8Fits:
    df_alD6D8[fit] = getBestRow(globals()[fit])
alDF = pd.DataFrame()
for fit in alphaFits:
   alDF[fit] = getBestRow(globals()[fit])
D10DF = pd.DataFrame()
for fit in D10Fits:
    D10DF[fit] = getBestRow(globals()[fit])
D6D8DF = pd.DataFrame()
for fit in D6D8Fits:
    D6D8DF[fit] = getBestRow(globals()[fit])
```

```python
plt.title(r'$\alpha_s(m_\tau)$ for the best fits of $\omega_\tau, \omega_{cube}$ and $\omega_{quartic}$')
plt.errorbar(list(alDF), alDF.loc['alpha'], alDF.loc['alphaErr'], fmt='.', marker='o', label=r'$\alpha_s(m_\tau)$')
plt.legend()
plt.xticks(rotation='vertical', size=16)
plt.yticks(size=16)
plt.gcf().set_size_inches(12, 6)
plt.tight_layout()
plt.gcf().savefig('../../Thesis/chapters/tauDecaysIntoHadrons/images/comparisonAlpha.png',dpi=300)
plt.show()
```

```python
print('mean: ', alDF.loc['alpha'].mean(), 'var: ', alDF.loc['alpha'].var())
```

In the upper plot we compare the best fits (chosen by $\chi^2/dof$ closest to $1$). The mean value and variance are given by $\mu_{\alpha_s}=0.3344$ and $\sigma^2_\alpha = 2.17e-5$. The errors of $\omega_{Quartic}$ for two and four OPE coefficients included are too small. Consequently these two fits should be excluded from the analysis. The rest of the fits are promising.

```python
plt.title(r'$c_6$ and $c_8$ for the best fits of $\omega_\tau, \omega_{cube}$ and $\omega_{quartic}$')
plt.errorbar(list(D6D8DF), D6D8DF.loc['c6'], D6D8DF.loc['c6Err'], fmt='.', marker='o', label=r'$c_6$')
plt.errorbar(list(D6D8DF), D6D8DF.loc['c8'], D6D8DF.loc['c8Err'], fmt='.', marker='o', label=r'$c_8$')
plt.ylim(-2.5, 1)
plt.legend()
plt.yticks(size=16)
plt.xticks(rotation='vertical', size=16)
plt.gcf().set_size_inches(12, 6)
plt.tight_layout()
plt.gcf().savefig('../../Thesis/chapters/tauDecaysIntoHadrons/images/comparisonD6D8.png', dpi=300)
plt.show()
```

In this plot we compare the $c_6$ and $c_8$ OPE coefficients of the best fits by $\chi^2/dof$ value. Even if the coefficients show a huge variation of value we present their mean and variance: $\mu_{c_6}=-0.252$ with $\sigma^2_{c_6}=0.29$ and $\mu_{c_8}=0.077$ with $\sigma^2_{c_8}=2.01$. We see that $c_6$ takes a negative and $c_{8}$ a postive value for most of the fits. Furthermore due to the variance of the OPE coefficients it is dificult to choose the right value for them. As there are more problems with including higher dimension contributions of the OPE we do not include them in our comparison.
