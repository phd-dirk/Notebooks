# Compare Kinematic Weight with wCube

```python
%store -r bestWKinematicRows
%store -r bestWCubeRows
bestRows = np.concatenate((bestWKinematicRows, bestWCubeRows))
c6s = bestRows[:,5]
c6Errs = bestRows[:,6]
c8s = bestRows[:,7]
c8Errs = bestRows[:,8]
chiDofs = bestRows[:,15]
x = np.array(['wK{D6,D8}', 'wK{D6,D8,Al}', 'wC{D6,D8}', 'wC{D6,D8,D10}', 'wC{D6,D8,D10,D12}', 'wC{D6,D8,alpha}'])
plt.subplot(3, 1, 1)
plt.title('c6s with errors')
plt.errorbar(x, c6s, yerr=c6Errs, barsabove=True, fmt='.', marker='o')
plt.subplot(3, 1, 2)
plt.title('c8s with errors')
plt.errorbar(x, c8s, yerr=c8Errs, barsabove=True, fmt='.', marker='o')
plt.subplot(3, 1, 3)
plt.title('Chi2/Dof')
plt.plot(chiDofs, marker='o', linestyle='None')
plt.show()
```

We are comparing the values for D6 and D8 for the kinematic weight (wK) with the
cubic weight (wC) for different parameter set ups ( e.g.. kinematic weight with
D6 and D8 free is expressed as "wK{D6,D8}"). The value for wK differ internaly,
if freeing alpha_s c8 and c6 are changing sign. We can also see that wC has
lower errors.
